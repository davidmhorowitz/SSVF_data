```python
#I'm going to manually look at every question.
#I don't think there's any way around it.
#To make it easy, I'm going to bundle questions
#with the same label together.

q_num_list = []

for q_num in dlp_qs["given_q_label"]: #for each question label
    if q_num in q_num_list:
        continue #if I already appended the question label (below), skip
    else:
        qs_per_num = ( #create a Series of questions for each q_num
            dlp_qs[
            (dlp_qs["given_q_label"] == q_num) #filter to only include the given q number
            &
            (
                (dlp_qs["source_type"] == "pdf")
            |
                (dlp_qs["source_type"] == "data dictionary")
            )
                #filter to only include pdfs and the data dictionary (since only those have question text)
        ]
        [["given_q_text", "num"]] #show the question text and their numbers
        )
        number_of_qs_per_num = len(qs_per_num)
        q_num_list.append(q_num)
        print(f'{q_num} ({number_of_qs_per_num} questions): {qs_per_num}')
        print("=======")
```

```python
#RESULTS FOR QS WITH THE SAME LABEL BUT DIFFERENT TEXT VALUES (DEFINITIONS):

#Q1: A presumably insignificant change in 21-22, from "this ssvf agency" to "this supportive services provider."

#Q2: There's an inconsistency. In the pdfs for 16 and for 17, Q2 asks if the participant would
#recommend the services provider to a friend or veteran. However, in the pdf for 21-22 and the data dictionary, it
#changes and asks if the SSVF case manager involved the participant in creating an individualized housing
#stabilization plan. Also, in the data dictionary, it refers to the "supportive services provider" rather than
#the "SSVF case manager." I would think they're referring to the same thing, but in my head, they mean different
#things.

#Q4: There's an inconsistency. In the pdfs for 16 and for 17, Q4 asks if the participant would return to
#the supportive services provider at no cost if she/he/they need help again. But in the pdf for 21-22 as well as 
#the data dictionary, Q4 instead asks if the SSVF case manager/supportive services provider involved the
#participant in creating an individualized housing stabilization plan. (The 20-22 pdf refers to the subject
#as "this SSVF agency," whereas the data dict refers to the subject as "the supportive services provider."
#I presume this is an insignificant change.)

#Q7: There's an inconsistency. In the pdfs for 16 and for 17, it asks for additional feedback about the supportive
#services provider. However, in the pdfs for 18, 20 and 21-22 as well as in the data dictionary, it instead asks
#how satisfied the participant is with the courteousness of the staff dealt with most often.

#Q8: There's an inconsistency. The pdf for 16 asks about supportive services received and their quality, whereas 
#the pdfs for 18, 20 and 21-22 are about timeliness of communication of the person the participant dealt with the most.

#Q9, Q10 and Q11: Appears insignificant, but just noting that the pdf for 21-22 has unusual prose. Presumably,
#these are all the same, though.

#QA: Presumably insignificant, but noting that for the pdf for 21-22, it asks for the number of people in the 
#household, whereas in the data dictionary, it asks for the number of individuals receiving support services 
#from the provider.
```

```python
#Redoing it with normalized letters (uppercasing them):

q_num_list = []

for q_num in dlp_qs["given_q_label"].str.upper(): #for each question label
    if q_num in q_num_list:
        continue #if I already appended the question label (below), skip
    else:
        qs_per_num = ( #create a Series of questions for each q_num
            dlp_qs[
            (dlp_qs["given_q_label"].str.upper() == q_num) #filter to only include the given q number
            &
            (
                (dlp_qs["source_type"] == "pdf")
            |
                (dlp_qs["source_type"] == "data dictionary")
            )
                #filter to only include pdfs and the data dictionary (since only those have question text)
        ]
        [["given_q_text", "num"]] #show the question text and their numbers
        )
        number_of_qs_per_num = len(qs_per_num)
        q_num_list.append(q_num)
        print(f'{q_num} ({number_of_qs_per_num} questions): {qs_per_num}')
        print("=======")
```

```python
#However, let's figure out whether we need to normalize the labels when accounting for *all* of the labels
#  not just the ones we filtered for in the loop:
```

```python
not_normalized_label_counts = (
    dlp_qs["given_q_label"]
    .value_counts()
    .to_frame()
    .reset_index()
)
not_normalized_label_counts.head()
```

```python
normalized_label_counts = (
    dlp_qs["given_q_label"]
    .str.upper()
    .value_counts()
    .to_frame()
    .reset_index()
)
normalized_label_counts.head()
```

```python
(
    not_normalized_label_counts
    ["given_q_label"] 
    == 
    normalized_label_counts
    ["given_q_label"]
).value_counts()
#Looks like there are some values that are different when uppercased.
#I'll have to stay wary of that.
```

```python
dlp_qs = (
    dlp_qs
    .merge(not_normalized_label_counts)
    .rename(columns={"count" : "not_normalized_label_counts"})
)
dlp_qs.head()
```

```python
#To add the count to a DataFrame:
counted_dlp_qs = (
    normalized_label_counts
    .assign(normalized_dlp_label = ( #this will be the key column they'll merge on
        lambda df: (
            df["given_q_label"] #defining normalized_dlp label as given_q_label
        ))) #end of assignment
    .drop(columns=["given_q_label"]) #dropping the given_q_label to only keep the count
    .merge #Now, they're merge on normalized_dlp_label
(
    dlp_qs
    .assign(
    normalized_dlp_label = lambda df: (
        df["given_q_label"] #adding the same column to this table to ensure the dfs merge
        .str.upper() 
    ) #end of this column's assignment
    ) #end of assign function
    ,
    on="normalized_dlp_label"
)
    .rename(columns={"count" : "normalized_label_counts"})
    [[ #reorganizing the columns a bit
        "given_q_label",
        "given_q_text",
        "source_type",
        "num",
        "percent_nonnull",
        "consistent/inconsistent",
        "dlp_std_q",
        "dlp_id_for_std_q",
        "dlp_qid_all_qs (index)",
        "notes",
        "given_q_text",
        "normalized_dlp_label",
        "normalized_label_counts",
        "not_normalized_label_counts"
        
    ]]
)
```

```python
counted_dlp_qs.head()
```

```python
#I want to know which Q labels with a source_type of "table" don't have a Q label with a source type of "pdf" or
#a source type of "data dictionary". I especially want to know this for observations where percent_nonnull is
#not na and is > 0.

#For this, I'll check if every given_q_label value with a source type of "pdf" or a source
#type of "data dictionary" appears in any given_q_label with a source type of "table."

(
    counted_dlp_qs
    [counted_dlp_qs["source_type"] == "table"]
    ["given_q_label"] #for labels with the source type of table ...
    .isin( 
        counted_dlp_qs
        [(counted_dlp_qs["source_type"] == "pdf")
        |
        (counted_dlp_qs["source_type"] == "data dictionary")]
        ["given_q_label"]) #see if any value from a pdf or the data dictionary appears within them.
).value_counts()

#The false values are how many "pdf" and "data dictionary" values
#don't exist in the "table" column
```

```python
#Let's identify those false values:

filter_for_table_labels_defined_by_pdf_or_dict = (
    counted_dlp_qs
    [counted_dlp_qs["source_type"] == "table"]
    ["given_q_label"]
    .isin(
        counted_dlp_qs
        [(counted_dlp_qs["source_type"] == "pdf")
        |
        (counted_dlp_qs["source_type"] == "data dictionary")]
        ["given_q_label"])
)
filter_for_table_labels_defined_by_pdf_or_dict
```

```python
table_labels_defined_by_pdf_or_dict = (
    counted_dlp_qs[filter_for_table_labels_defined_by_pdf_or_dict]
)
```

```python
#Let's try to fix this ... Looks like the indexes don't match.
```

```python
index_of_filter_for_table_labels_defined_by_pdf_or_dict = (
    filter_for_table_labels_defined_by_pdf_or_dict
    .index) #create an index for the filter

(
    counted_dlp_qs.loc[ #Within the main table,
        index_of_filter_for_table_labels_defined_by_pdf_or_dict, :] #locate rows with the index of the filter.
    [filter_for_table_labels_defined_by_pdf_or_dict] #Use that filter, as well.
)
```

```python
#I actually need the reverse because I need the True values
#to be the ones that don't include a source_type of "pdf"
#or "data dictionary." 

#Let's recreate the index with a ~ and use a ~
#in the final filter to get the reverse:
```

```python
reversed_filter_for_table_labels_defined_by_pdf_or_dict = ~filter_for_table_labels_defined_by_pdf_or_dict
```

```python
reversed_index_of_filter_for_table_labels_defined_by_pdf_or_dict = (
    reversed_filter_for_table_labels_defined_by_pdf_or_dict
    .index
) #create an index for the filter

cols_without_definitions_not_normalized = (
    counted_dlp_qs.loc[ #Again, within the main table,
        reversed_index_of_filter_for_table_labels_defined_by_pdf_or_dict, :] #locate rows with the index of the filter.
    [reversed_filter_for_table_labels_defined_by_pdf_or_dict] #Reverse the main filter, as well.
)
```
